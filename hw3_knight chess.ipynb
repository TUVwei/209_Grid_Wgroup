```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
actions = [[-2,-1],[-2,1],[2,-1],[2,1],[1,2],[1,-2],[-1,2],[-1,-2]]
```

```python
adj = np.zeros([64,64])
for s in range(64):
    for a in actions:
        i = int(s/8)
        j = s%8
        x = i + a[0]
        y = j + a[1]
        if 0<=x<8 and 0<=y<8:
            adj[s,8*x+y] = 1
            
```

```python
def bdf(mat_o, ini_node,end_node):
    # init
    mat=mat_o.copy()
    bfs = []
    tmp1_node = [ini_node]
    tmp2_node = []
    # begin
    while len(tmp1_node) + len(tmp2_node) > 0:
        # for each parent
        for parent_node in tmp1_node:
            tmp = [parent_node, []]
            # for each child, row
            for child_node, child_weight in enumerate(mat[parent_node, :]):
                if child_weight > 0:
                    # add current node info
                    tmp[1].append(child_node)
                    # add children to tmp2
                    tmp2_node.append(child_node)
                    # remove this edge
                    mat[parent_node, child_node] = 0
            # for each child, col
            for child_node, child_weight in enumerate(mat[:, parent_node]):
                if child_weight > 0:
                    # add current node info
                   # tmp[1].append(child_node)
                    #add child node
                   # tmp2_node.append(child_node)
                    # remove this edge
                    mat[child_node, parent_node] = 0
            # bfs add
            bfs.append(tmp.copy())
        # to next level
        tmp1_node = tmp2_node.copy()
        tmp2_node = []
    keys=[]
    values=[]
    for i in range(len(bfs)):
        keys.append(bfs[i][0])
        values.append(bfs[i][1])
    bfs_dict=dict(zip(keys, values))        
    queue = []
    # push the first path into the queue
    queue.append([ini_node])
    while queue:
        # get the first path from the queue
        path = queue.pop(0)
        # get the last node from the path
        node = path[-1]
        # path found
        if node == end_node:
            break
        # enumerate all adjacent nodes, construct a 
        # new path and push it into the queue
        for adjacent in bfs_dict[node]:
            new_path = list(path)
            new_path.append(adjacent)
            queue.append(new_path)
    return bfs,path
```

```python
bfs,path = bdf(adj,1,10)
```
